# Visualização dos passes de um determinado jogador

## Importação das bibliotecas

```python
import pandas as pd
from statsbombpy import sb
from mplsoccer.pitch import Pitch
```

## Carregamento dos dados

```python
comp = sb.competitions()
comp
```

```python
matches = sb.matches(11, 90)
matches
```

```python
events = sb.events(3773372, split=True, flatten_attrs=True)['passes']
```

```python
events.columns
```

```python
events.pass_outcome.unique()
```

## Preparação dos Dados

```python
team1 = 'Barcelona'
team2 = 'Atlético Madrid'
events = events[(events.possession_team == team1) & (events.pass_outcome.isnull())]
```

```python
events.pass_end_location
```

```python
x = [x[0] for x in events.location]
y = [y[1] for y in events.location]
x_end = [x_end[0] for x_end in events.pass_end_location]
y_end = [y_end[1] for y_end in events.pass_end_location]

events = events.assign(x=x, y=y, x_end=x_end, y_end=y_end)
events = events.drop(['location', 'pass_end_location'], axis=1)

events.head()
```

```python
events.pass_outcome.unique()
```

## Gráfico

```python
pitch = Pitch(pitch_type='statsbomb', line_zorder=2, line_color='lightgray')
bins = (5, 4)
fig, ax = pitch.draw(figsize=(16, 10), constrained_layout=True, tight_layout=True)
fig.set_facecolor('lightgray')
pitch_bins = pitch.bin_statistic(events.x, events.y, statistic='count', bins=bins)
pitch_heatmap = pitch.heatmap(pitch_bins, ax=ax, cmap='YlOrRd', edgecolor='black')
arrows = pitch.flow(events.x, events.y, events.x_end, events.y_end, color='black', arrow_type='same', 
                   arrow_length=8, bins=bins, ax=ax)
#title = ax.set_title("", fontweight='bold')
```
