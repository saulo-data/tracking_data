# Passes do Neymar na partida contra a Bélgica pela Copa 2018

## Importação das Bibliotecas

```python
from statsbombpy import sb
from mplsoccer.pitch import Pitch
```

## Preparação e limpeza dos dados

```python
comp = sb.competitions()
comp.head()
```

```python
matches = sb.matches(43, 3)
events = sb.events(match_id=8650)
events.tail()
```

```python
passes = events[events.type == 'Pass']
passes_ney = passes[passes.player == 'Neymar da Silva Santos Junior']
passes_ney = passes_ney[passes_ney.pass_outcome != 'Incomplete']
passes_ney_final = passes_ney[['player', 'location', 'pass_end_location', 
                               'pass_recipient']]
```

```python
passes_ney_final
```

## Preparação das coordenadas dos passes

```python
x = [x[0] for x in passes_ney_final.location]
y = [y[1] for y in passes_ney_final.location]
x_end = [x[0] for x in passes_ney_final.pass_end_location]
y_end = [y[1] for y in passes_ney_final.pass_end_location]
```

```python
passes_ney_final = passes_ney_final.assign(x=x, y=y, x_end=x_end, y_end=y_end)
```

```python
passes_ney_final
```

```python
passes_ney_final = passes_ney_final.drop(['location', 'pass_end_location'], axis=1)
```

```python
passes_ney_final
```

```python
average_locations = passes_ney_final.groupby('pass_recipient').agg({'x_end':['mean'],'y_end':['mean', 'count']})
```

```python
average_locations.columns = ['x_end','y_end','count']
average_locations['x'] = sum(passes_ney_final['x']) / len(passes_ney_final['x'])
average_locations['y'] = sum(passes_ney_final['y']) / len(passes_ney_final['y'])
```

```python
average_locations
```

## Gráficos

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white', 
              figsize=(20, 10), constrained_layout=True, tight_layout=False)
fig, ax = pitch.draw()

ax.set_title("Passes do Neymar na partida")

pitch.scatter(average_locations.x,average_locations.y,
                      s=80, color='yellow', edgecolors='black',
                      linewidth=1, alpha=1, zorder=1, ax=ax)
pitch.scatter(average_locations.x_end,average_locations.y_end,
                      s=average_locations['count']*6, color='blue', 
                      edgecolors='black', linewidth=1, alpha=1, zorder=1, 
                      ax=ax)
pitch.arrows(average_locations.x, average_locations.y, 
                      average_locations.x_end, average_locations.y_end,
                      width=1.5, color='white', ax=ax, zorder=1, alpha=.6);
```
